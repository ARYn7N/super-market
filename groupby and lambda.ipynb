```python
import pandas as pd

# Sample data
data = {
    'Store': ['Store A', 'Store A', 'Store B', 'Store B', 'Store C', 'Store C'],
    'Product': ['Product 1', 'Product 2', 'Product 1', 'Product 2', 'Product 1', 'Product 2'],
    'Sales': [100, 150, 200, 250, 300, 350]
}

# Create DataFrame
df = pd.DataFrame(data)
```

```python
grouped_df = df.groupby('Store')['Sales'].sum().reset_index()
print(grouped_df)
```
