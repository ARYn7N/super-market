```python
sns.countplot(data=df, x='Gender')
plt.title('Customer Gender Distribution')
plt.show()
```

```python
print(df['Gender'].value_counts())
```

```python
sns.histplot(df['Rating'], kde=True)
plt.title('Customer Rating Distribution')
plt.show()
```

```python
print("Skewness:", df['Rating'].skew())
```

```python
branch_sales = df.groupby('Branch')['Total'].sum().sort_values(ascending=False)
branch_sales.plot(kind='bar', title='Total Sales by Branch')
plt.ylabel('Sales')
plt.show()
```

```python
sns.countplot(data=df, x='Payment')
plt.title('Payment Method Distribution')
plt.show()
```

```python
branch_payment = pd.crosstab(df['Branch'], df['Payment'])
branch_payment.plot(kind='bar', stacked=True, title='Branch-wise Payment Distribution')
plt.show()
```

```python
# Drop missing values to avoid plotting errors
df_clean = df.dropna(subset=['gross income', 'Rating'])

# Create scatter plot
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df_clean['gross income'], y=df_clean['Rating'], hue=df_clean['Branch'], palette='Set2')
plt.title('gross Income vs Customer Rating')
plt.xlabel('gross Income')
plt.ylabel('Customer Rating')
plt.grid(True)
plt.show()
```

```python
sns.boxplot(data=df, x='Branch', y='Rating')
plt.title('Rating Distribution by Branch')
plt.show()
```

```python
profit_by_branch = df.groupby('Branch')['gross income'].sum().sort_values(ascending=False)
profit_by_branch.plot(kind='bar', title='Profit by Branch')
plt.ylabel('gross income')
plt.show()
```

```python
sns.boxplot(data=df, x='Gender', y='gross income')
plt.title('Gross Income by Gender')
plt.show()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
daily_income = df.groupby('Date')['gross income'].sum()
daily_income.plot(title='Daily Gross Income Trend')
plt.ylabel('gross income')
plt.show()
```

```python
income_by_product = df.groupby('Product line')['gross income'].sum().sort_values(ascending=False)
income_by_product.plot(kind='bar', title='Income by Product Line')
plt.ylabel('gross income')
plt.show()
```
